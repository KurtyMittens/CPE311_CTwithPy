### Years in the Data Set

- Now let's check the growth of the price of the per year of the Price per square Feet

```python
years_of_built = np.sort(housing_price_df['YearBuilt'].unique()) # Getting all the years that indicated in the First Data Frame (sorted)
```

```python
years_of_built # Printing all of the dates
```

- with this years, we can Forecast the growth of the prices using the 72 years of data set

### Growth Of the Price / SquareFeet in Rural Houses in 72 Years

- Let's Create a Data Frame for the Growth of price in Rural Houses

```python
rural_growth_df = pd.DataFrame(years_of_built, columns = ["Year"]) # Creating Another Data Frame for forecast of the Price / Square Feet
```

```python
rural_growth_df["GrowthMeanPriceSq"] = rural_growth_df.Year.apply(lambda x: rural_price_df[rural_price_df["YearBuilt"] == x]["Price/SqFeet"].mean())
 # Getting the total Mean of the "Price/SqFeet" of a particular year
```

```python
rural_growth_df # The Rural Growth Price DataFrame in 72 years
```

### Growth Of the Price / SquareFeet in Suburbs Houses in 72 Years

- Let's Do it in Suburbs Houses Price Growth also

```python
suburbs_growth_df = pd.DataFrame(years_of_built, columns = ["Year"]) # Creating Another Data Frame for forecast of the Price / Square Feet
```

```python
suburbs_growth_df["GrowthMeanPriceSq"] = suburbs_growth_df.Year.apply(lambda x: suburb_price_df[suburb_price_df["YearBuilt"] == x]["Price/SqFeet"].mean())
# Getting the total Mean of the "Price/SqFeet" of a particular year
```

```python
suburbs_growth_df # The Suburbs Growth Price DataFrame in 72 years
```

#### Testing the Analysis (Demand of people)

**NOTE:**

"By the law of demand, as price decreases, the quantity of housing demanded increases."

```python
demand_df = pd.DataFrame(years_of_built, columns=["Year"])
```

```python
demand_df["Rural"] = rural_growth_df.GrowthMeanPriceSq.apply(lambda x:x)
```

```python
demand_df["Suburbs"] = suburbs_growth_df.GrowthMeanPriceSq.apply(lambda x:x)
```

```python
demand_df["Urban"] = urban_growth_df.GrowthMeanPriceSq.apply(lambda x:x)
```

```python
demand_df["Demand"] = demand_df.idxmin(axis=1)
```

```python
demand_df
```

```python
demand_df["Demand"].value_counts()
```

```python
plt.pie(demand_df["Demand"].value_counts(), labels=["Suburbs", "Rural", "Urban"],  autopct='%1.1f%%')
plt.title("Chart of DEMAND of HOUSING NIEGHBORHOOD base on Housing Prices for the past 72 Years")
plt.show()
```
